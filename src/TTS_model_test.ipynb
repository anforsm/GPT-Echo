```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("anforsm/distilgpt2-finetuned-common-voice")
tokenizer.pad_token_id = tokenizer.eos_token_id
def tokenize(text):
    prompt = f"text: {text}\nsound:"
    return tokenizer(prompt, return_tensors="pt")
```

```python
print(tokenize("Hello")["input_ids"][0])
[print(token - 2 ) for token in tokenize("Hello")["input_ids"][0] if token > 2]
model.forward(tokenize("Hello")["input_ids"])
```
